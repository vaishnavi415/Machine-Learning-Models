```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

## **ORIGINAL DATASET**

```python
name = 'mnist_784'
mnist = fetch_openml(name,cache=)
df = pd.DataFrame(mnist.data,columns=[i for i in range(0,784)])
df['Output'] = mnist.target.astype('int')
df_five = df[df['Output'] <= 4]
df_five
```

```python
train, validate, test = np.split(df_five.sample(frac=1, random_state=42), [int(.7*len(df_five)), int(.9*len(df_five))])
```

```python
train
```

```python
validate
```

```python
test
```

```python
X_train = train[train.columns[:-1]]
y_train = train[train.columns[-1]]

X_validate = validate[validate.columns[:-1]]
y_validate = validate[validate.columns[-1]]

X_test = test[test.columns[:-1]]
y_test = test[test.columns[-1]]
```

## **NEAREST NEIGBOUR MODEL**

```python
from sklearn.neighbors import KNeighborsClassifier as KNC
k = KNC().fit(X_train,y_train)
```

## **PERCEPTRON MODEL**

```python
from sklearn.linear_model import Perceptron
p = Perceptron().fit(X_train,y_train)
```

## **SVM MODEL**

```python
from sklearn.svm import SVC
s = SVC().fit(X_train,y_train)
```

## **ACCURACY OF ALL MODELS**

```python
y_pred_k = k.predict(X_test)
from sklearn.metrics import accuracy_score
acc_k = accuracy_score(y_pred_k, y_test)
print("THE ACCURACY OF NORMAL KNC ",acc_k)

y_pred_p = p.predict(X_test)
from sklearn.metrics import accuracy_score
acc_p = accuracy_score(y_pred_p, y_test)
print("THE ACCURACY OF NORMAL PERCEPTRON ",acc_p)

y_pred_s = s.predict(X_test)
from sklearn.metrics import accuracy_score
acc_s = accuracy_score(y_pred_s, y_test)
print("THE ACCURACY OF NORMAL SVM ",acc_s)
```

## **TUNED MODELS - KNC, PERCEPTRON,SVM**

```python
from sklearn.model_selection import GridSearchCV as GSCV


k_param = {'n_neighbors' : list(np.arange(2,6,1)), 'weights' : ('uniform','distance')}
k_tuned = GSCV(KNC(),k_param).fit(X_validate,y_validate)

p_param = {'penalty' : ('l1','l2'), 'tol' : [0.001 * 10 ** (n - 1) for n in range(1, 4 + 1)], 'class_weight' : (None,'balanced')}
p_tuned = GSCV(Perceptron(),p_param).fit(X_validate,y_validate)

s_param = {'kernel' : ('linear','poly','rbf'), 'C' : [0.8,1.0,1.2], 'tol' : [0.001 * 10 ** (n - 1) for n in range(1, 4 + 1)]}
s_tuned = GSCV(SVC(),s_param).fit(X_validate,y_validate)
```

```python
k_tuned.cv_results_
```

```python
p_tuned.cv_results_
```

```python
s_tuned.cv_results_
```

## **BEST ESTIMATOR OF TUNED MODELS**

```python
k_tuned_best = k_tuned.best_estimator_
print("THE BEST ESTIMATOR FOR KNC IS",'\n',k_tuned_best)

p_tuned_best = p_tuned.best_estimator_
print("THE BEST ESTIMATOR FOR PERCEPTRON IS",'\n',p_tuned_best)

s_tuned_best = s_tuned.best_estimator_
print("THE BEST ESTIMATOR FOR SVC IS",'\n',s_tuned_best)
```

## **ACCURACY OFTUNED MODELS**

```python
y_pred_k_tun = k_tuned_best.predict(X_test)
from sklearn.metrics import accuracy_score
acc_k_tun = accuracy_score(y_pred_k_tun, y_test)
print("THE ACCURACY OF TUNED KNC ",acc_k_tun)

y_pred_p_tun = p_tuned_best.predict(X_test)
from sklearn.metrics import accuracy_score
acc_p_tun = accuracy_score(y_pred_p_tun, y_test)
print("THE ACCURACY OF TUNED PERCEPTRON ",acc_p_tun)

y_pred_s_tun = s_tuned_best.predict(X_test)
from sklearn.metrics import accuracy_score
acc_s_tun = accuracy_score(y_pred_s_tun, y_test)
print("THE ACCURACY OF TUNED SVM ",acc_s_tun)
```
